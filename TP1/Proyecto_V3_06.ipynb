**5 Glosario de términos en el dataset**

La información contenida en el dataset abarca los siguientes puntos:

Color: Color del vehículo.

Dealer_zip_code: Código postal del anunciante.

Description: Descripción presente en el anuncio del vehículo.

Fuel: Tipo de combustible del vehículo (diésel, gasolina, eléctrico, híbrido).

Is_professional: Indica si el vendedor es profesional (un concesionario).

Km: Kilometraje del vehículo.

Location: Ciudad en la que se ha publicado el anuncio.

Make: Marca del vehículo.

Model: Modelo del vehículo.

Photos: Número de fotos del vehículo disponibles en el anuncio.

Power: Potencia del vehículo.

Price: Precio de venta del vehículo.

Publish_date: Fecha de publicación del anuncio.

Shift: Tipo de cambio (Automático/Manual).

Update_date: Fecha de actualización del anuncio.

Vehicle_type: Tipo de vehículo: coche, moto...

Version: Versión del vehículo.

**Gráfica de todas las marcas con su precio**

**La gráfica no es estética y la información confunde mucho por esta razón más abajo se seleccionan solo las primeras marcas más caras**

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(df_vehiculos['make'], df_vehiculos['price'])
ax.set_title('Marcas y precio')
ax.set_ylabel('Precio-Euros')
ax.set_xlabel('Marca')
```

**Para mostrar solo las primeras 100 filas con los precios MÁS ALTOS, se ordenó el DataFrame por precio y luego seleccionaron las primeras 100 filas y se observa en gráfica de barra que en estas 100 filas aparecen 6 marcas.**

```python
# Ordenar el DataFrame por precio en orden descendente
df_vehiculos_ordenado = df_vehiculos.sort_values(by='price', ascending=False)

# Seleccionar las primeras 10 marcas con el precio más alto
primeras_10_marcas = df_vehiculos_ordenado.head(100)

# Crear el gráfico
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(primeras_10_marcas['make'], primeras_10_marcas['price'])
ax.set_title('Top 10 Marcas con el Precio más Alto')
ax.set_ylabel('Precio')
ax.set_xlabel('Marca')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Para mostrar solo las primeras 100 filas con los precios MÁS BAJOS, se ordenó el DataFrame por precio del más económico al más caro; luego se seleccionaron las primeras 100 filas y se observa en gráfica de barras que en estas 100 filas aparecen 16 marcas.**

**Con el código anterior solo se cambio la palabra TRUE en ASCENDING, para indicar que será de menor a mayor.**

```python
# Ordenar el DataFrame por precio en orden ASCENDENTE
df_vehiculos_ordenado = df_vehiculos.sort_values(by='price', ascending=True)

# Seleccionar las primeras 10 marcas con el precio más bajo
primeras_10_marcas = df_vehiculos_ordenado.head(100)

# Crear el gráfico
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(primeras_10_marcas['make'], primeras_10_marcas['price'])
ax.set_title('Top 10 Marcas con el Precio más Bajos')
ax.set_ylabel('Precio')
ax.set_xlabel('Marca')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Se crea grupo de marcas con el número de veces que se repite con la intensión de entender cuales son las marcas más vendidas**

```python
# Método groupby
#carros=
df_vehiculos.groupby('make').size().head(30)
#print(df_vehiculos.groupby('make')['price'].head(15))
```

**Se agrupan las marcas y se muestra en una gráfica de barras las 10 más numerosas**

```python
# Primero, agrupamos los datos por marca y contamos el número de vehículos por marca
vehiculos_por_marca = df_vehiculos.groupby('make').size()

# Seleccionamos las primeras 10 marcas más numerosas
top_10_marcas = vehiculos_por_marca.nlargest(10)

# Graficamos los daos
plt.figure(figsize=(10, 6))
top_10_marcas.plot(kind='bar', color='skyblue')
plt.title('Top 10 marcas de vehículos más numerosas')
plt.xlabel('Marca')
plt.ylabel('Número de vehículos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
